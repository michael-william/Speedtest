```python
df = pd.read_csv(r'/users/michaelcondon/Desktop/speedtest/speedtest_logs.csv', parse_dates=[3,4])
df.rename(columns={'time':'hour'}, inplace=True)
#df['time'] = pd.to_datetime(df.time, format='%H:%M:%S')
#df['time'] = [time.time() for time in df['time']]
df['hour'] = df.hour.dt.round('H').dt.hour
df['week'] = df.date.dt.week
df['month'] = df.date.dt.month
```

```python
df
```

```python
ping = go.Bar(
    x=df['hour'],
    y=df['ping'],
    opacity = 0.75,
    hoverinfo="name + y",
    name='Ping(ms)')

down = go.Bar(
    x=df['hour'],
    y=df['download'],
    opacity = 0.75,
    hoverinfo="name + y",
    name='Download (mbs)')

up = go.Bar(
    x=df['hour'],
    y=df['upload'],
    opacity = 0.75,
    hoverinfo="name + y",
    name = 'Upload (mbs)')

data = [down, up, ping]

layout = go.Layout(
    title='WiFi Stats',
    barmode='overlay', 
    hovermode='y')

fig = go.Figure(data=data, layout=layout)
fig.update_layout(hovermode='closest')
#pyo.plot(fig, filename='Wifi_Stats.html')
fig.show()
```
